5. Excluding Low-Activity Customers

```python
# Create Column for Max Value of orders_#
df_opc['max_order'] = df_opc.groupby(['user_id'])['orders_#'].transform(np.max)
```

```python
# Assigning Low Activity & Active
result = []

for value in df_opc["max_order"]:
  if value < 5:
    result.append("Low Activity")
  else:
    result.append("Active")
```

```python
df_opc['Customer_activity'] = result
```

```python
df_opc['Customer_activity'].value_counts()
```

```python
# Isolating Active Customers
opc_active = df_opc[df_opc['Customer_activity'] == 'Active']
```

```python
# Make Data Type Consistent Between Dataframes
df_depts['department_id'] = df_depts['department_id'].astype('int64')
df_opc_active['department_id'] = df_opc_active['department_id'].astype('int64')
```

7. Profiling Variables

```python
df_opcd.describe()
```

```python
# Create Age Group Categories
df_opcd.loc[(df_opcd['age'] >=18) & (df_opcd['age'] <35), 'age_group'] = 'Young Adult'
df_opcd.loc[(df_opcd['age'] >=35) & (df_opcd['age'] <65), 'age_group'] = 'Middle Aged'
df_opcd.loc[df_opcd['age'] >=65, 'age_group'] = 'Senior'


df_opcd['age_group'].value_counts(dropna = False)
```

```python
df_opcd.head()
```

```python
# Create Income Categories
df_opcd.loc[df_opcd['income'] < 70000, 'income_group'] = 'Lower Income'
df_opcd.loc[(df_opcd['income'] >= 70000) & (df_opcd['income'] < 130000), 'income_group'] = 'Average Income'
df_opcd.loc[df_opcd['income'] >= 130000, 'income_group'] = 'High Income'


df_opcd['income_group'].value_counts(dropna = False)
```

```python
# Create Family Types
df_opcd.loc[(df_opcd['income_group'] == 'Lower Income') & (df_opcd['n_dependants'] == 0), 'family_type'] = 'Lower Income No Kids'
df_opcd.loc[(df_opcd['income_group'] == 'Lower Income') & (df_opcd['n_dependants'] >= 1), 'family_type'] = 'Lower Income W/ Kids'
df_opcd.loc[(df_opcd['income_group'] == 'Average Income') & (df_opcd['n_dependants'] == 0), 'family_type'] = 'Average Income No Kids'
df_opcd.loc[(df_opcd['income_group'] == 'Average Income') & (df_opcd['n_dependants'] >= 1), 'family_type'] = 'Average Income W/ Kids'
df_opcd.loc[(df_opcd['income_group'] == 'High Income') & (df_opcd['n_dependants'] == 0), 'family_type'] = 'High Income No Kids'
df_opcd.loc[(df_opcd['income_group'] == 'High Income') & (df_opcd['n_dependants'] >= 1), 'family_type'] = 'High Income W/ Kids'

df_opcd['family_type'].value_counts(dropna = False) 
```

10. Analyzing Profile Expenses by Department

```python
# Create Bar Chart for Age Groups, Departments, & Prices
plt.figure(figsize=(10, 10))

sns.barplot(data = df_opcd, 
             x='prices',          
             y = 'department', 
             hue='age_group', 
             palette = 'viridis',
             estimator= np.mean,
             ci=None
            )

plt.title("Department Expenditures by Age Group")
plt.xlabel("Total Revenue")
plt.ylabel("")
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_group_dept_expenses.png'))
plt.show()
```

```python
# Create Bar Chart for Income Groups, Departments, & Prices
plt.figure(figsize=(10, 10))

sns.barplot(data = df_opcd, 
             x='prices',          
             y = 'department', 
             hue='income_group', 
             palette = 'rocket',
             estimator= np.mean,
             ci=None
            )

plt.title("Department Expenditures by Income Group")
plt.xlabel("Total Revenue")
plt.ylabel("")
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'income_groups_dept_expenses.png'))
plt.show()
```

```python
# Create Bar Chart for Family Type, Departments, & Prices
plt.figure(figsize=(10, 10))

sns.barplot(data = df_opcd, 
             x='prices',          
             y = 'department', 
             hue='family_type', 
             palette = 'cubehelix',
             estimator= np.mean,
             ci=None
            )

plt.title("Department Expenditures by Family Type")
plt.xlabel("Total Revenue")
plt.ylabel("")
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'fam_type_dept_expenses.png'))
plt.show()
```

Looking at the different profiles by department revealed few differences. It looks like the different profiles created showed similar behaviors in the different departments. The final profile, Family Type, did show some interesting behavior of the higher income folks, who have a significant impact on the dairy and egg purchases. Higher income folks purchase a lot more from the dairy and eggs department -- it would make sense to target dairy and egg marketing more directly at higher income folks.

Age groups did not reveal much when analyzed by region. Family type showed the highest earners to spend the most, but showed that having kids would not mean that the family would spend more. Interestingly, in the lower income group, having kids appears to have no impact on expenditures.
