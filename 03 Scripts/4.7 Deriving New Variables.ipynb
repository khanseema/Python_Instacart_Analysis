3. Assign Price Ranges

```python
# Create Subset
df = df_ords_prods_merged[:1000000]
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply price_range to DF
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

4. Create Busiest Day Column

```python
# Determine Busiest and Least Busy Days
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Assign String Values to Days
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create New Column for Busiest Days
df_ords_prods_merged['busiest_day'] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

5. Create New Column for Busiest Days

```python
# Assign String Values to Days
result2 = []

for value in df_ords_prods_merged['orders_day_of_week']: 
    if value == 0:
        result2.append('Busiest days')
    elif value == 1:
        result2.append('Busiest days')
    elif value == 4: 
        result2.append('Slowest days')
    elif value == 3:
        result2.append('Slowest days')
    else:
        result2.append('Regular days')
```

```python
# Create New Column for Busiest Days (Plural)
df_ords_prods_merged['busiest_days'] = result2
```

```python
df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

```python
# Assign String Values to Hours (most orders = > 2,000,000 & fewest orders = < 100,000)
result3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 10:
    result3.append("Most orders")
  elif value == 11:
    result3.append("Most orders")
  elif value == 14:
    result3.append("Most orders")
  elif value == 15:
    result3.append("Most orders")
  elif value == 13:
    result3.append("Most orders")
  elif value == 12:
    result3.append("Most orders")
  elif value == 16:
    result3.append("Most orders")
  elif value == 9:
    result3.append("Most orders")
  elif value == 17:
    result3.append("Most orders")
  elif value == 5:
    result3.append("Fewest orders")
  elif value == 2:
    result3.append("Fewest orders")
  elif value == 4:
    result3.append("Fewest orders")
  elif value == 3:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```
