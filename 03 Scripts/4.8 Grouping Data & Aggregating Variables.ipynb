```python
# Import Dataframes
df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_derived_cols.pkl'))
```

```python
# Aggregate Mean of order_number by department_id
df.groupby('department_id').agg({'orders_#': ['mean']})
```

```python
# Aggregate Mean of order_number by department_id Full Dataframe Q2
df_ords_prods.groupby('department_id').agg({'orders_#': ['mean']})
```

Q3 - In the subset, only a small number of departments is represented. Also, the mean values for the departments included in the subset changed slightly with the full dataframe.

```python
df_ords_prods['max_order'] = df_ords_prods.groupby(['user_id'])['orders_#'].transform(np.max)
```

```python
df_ords_prods.head(15)
```

```python
df_ords_prods.loc[df_ords_prods['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods.loc[(df_ords_prods['max_order'] <= 40) & (df_ords_prods['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods.loc[df_ords_prods['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods['loyalty_flag'].value_counts(dropna = False)
```

```python
# Aggregate Stats of prices by loyalty_flag Q5
df_ords_prods.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Q5 - It appears that it is the new customers that have the highest average price product.

```python
# Create Column with Price Mean by User Q6
df_ords_prods['mean_price'] = df_ords_prods.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods.head(15)
```

```python
df_ords_prods.loc[df_ords_prods['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods.loc[df_ords_prods['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods['spending_flag'].value_counts(dropna = False)
```

```python
# Create Column with days_since_prior_order Median by User Q7
df_ords_prods['median_buying_frequency'] = df_ords_prods.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods.head(15)
```

```python
df_ords_prods.loc[df_ords_prods['median_buying_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods.loc[(df_ords_prods['median_buying_frequency'] <= 20) & (df_ords_prods['median_buying_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods.loc[df_ords_prods['median_buying_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods['order_frequency_flag'].value_counts(dropna = False)
```
