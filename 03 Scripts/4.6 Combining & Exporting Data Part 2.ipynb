4.6 Combining & Exporting Data Part 2

1. Import Dataframes

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

2. Import Dataframes

```python
# Create String of Path to Main Project Folder
path = r'/Users/seema/Documents/CareerFoundry/A4/Instacart Basket Analysis'
```

```python
# Import Dataframes
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
df_ords_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

3. Check Data

```python
# Check Output
df_prods.head()
```

```python
# Check Output
df_ords_combined.head()
```

```python
# Drop Unnamed Columns
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
# Drop Unnamed Columns
df_ords_combined = df_ords_combined.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'])
```

```python
# Drop _merge Column
df_ords_combined = df_ords_combined.drop(columns = ['_merge'])
```

4. Merge Data

```python
# Create Merged Dataframe
df_ords_prods_merged = df_ords_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
# Check Output & Confirm Merge
df_ords_prods_merged.head()
```

```python
#Checking Values
df_ords_prods_merged['_merge'].value_counts()
```

5. Export to pkl

```python
# Export Data to pkl
df_ords_prods_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
